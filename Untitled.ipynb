```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split,cross_val_score,cross_val_predict
from sklearn.metrics import accuracy_score,f1_score,precision_score,confusion_matrix,recall_score,mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor,AdaBoostRegressor,BaggingRegressor,GradientBoostingRegressor,ExtraTreesRegressor
from sklearn.tree import DecisionTreeRegressor
```

```python
df = pd.read_csv("kc_house_data.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull
```

```python
df.drop('id',axis = 1,inplace= True)
```

```python
df.drop('date',axis = 1,inplace= True)
```

```python
df.corr()
```

```python
housing = df.astype(np.int64)
```

```python
housing.corr()
```

```python
housing.dtypes
```

```python
housing.hist(figsize = (20,20))
```

```python
x = housing.drop('price',axis = 1)
```

```python
y = housing.get('price')
```

```python
x.head()
```

```python
y.head()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
models = []
# models.append(('LR',LinearRegression()))
models.append(('Ex',ExtraTreesRegressor()))
models.append(('Dt',DecisionTreeRegressor()))
# models.append(('AR',AdaBoostRegressor()))
# models.append(('RR',RandomForestRegressor()))
# models.append(('BR',BaggingRegressor()))
# models.append(('GR',GradientBoostingRegressor()))
models
```

```python
results = []
name = []
```

```python
for name, model in models:
    model.fit(x_train,y_train)
    results.append(model.score(x_train,y_train))
    results
        
```

```python
print(results)
```

```python
for name, model in models:
    model.fit(x_train,y_train)
    
#     obj = model.predict(x_test)
    results.append(model.score(x_test,y_test))
    print(results)
```

```python
model.predict(x_test)
```

```python
y_train_pred = cross_val_score(model,x_test,y_test,cv=3,scoring='accuracy')
y_train_pred
```

```python
y_test_pred = cross_val_predict(model,x_test,y_test,cv=3)
y_test_pred
```

```python
y_pred = model.predict(x_test)
pd.crosstab(y_test,y_pred,rownames=['True'], colnames=['Predicted'], margins=True)
```

```python
mse = mean_squared_error(y_test,y_test_pred)
```
